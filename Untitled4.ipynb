```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import sklearn 
from sklearn.preprocessing import LabelEncoder 
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest , mutual_info_classif
```

```python
df_imbalanced = pd.read_csv('US_Chronic_Disease_Indicators.csv')
```

```python
# Removing useless features 
df_imbalanced = df_imbalanced.drop('StratificationCategory3' ,axis = 1)
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest, chi2, RFE
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder

# Load the dataset
# Assuming your dataset is already loaded and named 'df_imbalanced'
# If not, you can load it using pd.read_csv('your_dataset.csv')
# df_imbalanced = pd.read_csv('your_dataset.csv')

# Now label encoding the remaining features
# Create a LabelEncoder object
label_encoder = LabelEncoder()

for col in df_imbalanced.keys():
    if df_imbalanced[col].dtype == 'object':
        df_imbalanced[col] = label_encoder.fit_transform(df_imbalanced[col].astype(str))

# Target column
target_column = 'Topic'
df_imbalanced = df_imbalanced.dropna()

# Separate features and target
X = df_imbalanced.drop(target_column, axis=1)
y = df_imbalanced[target_column]

# Correlation-based Feature Selection
corr_matrix = X.corr()
corr_with_target = corr_matrix[target_column].abs().sort_values(ascending=False)
threshold = 0.5
selected_features_corr = corr_with_target[corr_with_target > threshold].index.tolist()
print("Selected features using correlation method:", selected_features_corr)

# SelectKBest (using chi-square statistic)
best_features_selector = SelectKBest(score_func=chi2, k=5)  # Select top 5 features
best_features = best_features_selector.fit(X, y)
selected_features_kbest = X.columns[best_features.get_support()].tolist()
print("Selected features using SelectKBest method:", selected_features_kbest)

# Recursive Feature Elimination (RFE)
model = LogisticRegression(max_iter=1000)  # You can use another suitable model
rfe_selector = RFE(estimator=model, n_features_to_select=5, step=1)  # Select top 5 features
rfe_features = rfe_selector.fit(X, y)
selected_features_rfe = X.columns[rfe_features.support_].tolist()
print("Selected features using RFE method:", selected_features_rfe)

# Find common features
common_features = set(selected_features_corr) & set(selected_features_kbest) & set(selected_features_rfe)
print("Common features across methods:", list(common_features))
```

```python
# Calculate correlation matrix
corr_matrix = df_imbalanced.corr()

# Get absolute correlation values with the target column
corr_with_target = corr_matrix[target_column].abs().sort_values(ascending=False)

# Select features with correlation above a certain threshold (e.g., 0.5)
threshold = 0.2
selected_features_corr = corr_with_target[corr_with_target > threshold].index.tolist()
print("Selected features using correlation method:", selected_features_corr)
```

```python
# Separate features and target
X = df_imbalanced.drop('Topic' , axis = 1)
y = df_imbalanced['Topic']

# Apply SelectKBest
best_features_selector = SelectKBest(score_func=chi2, k=7)  # Select top 5 features
best_features = best_features_selector.fit(X, y)

# Get selected features
selected_features_kbest = X.columns[best_features.get_support()].tolist()
print("Selected features using SelectKBest method:", selected_features_kbest)
```

```python
# Initialize model for RFE
model = LogisticRegression(max_iter=100)  # You can use another suitable model

# Apply RFE
rfe_selector = RFE(estimator=model, n_features_to_select=5, step=1)  # Select top 5 features
rfe_features = rfe_selector.fit(X, y)

# Get selected features
selected_features_rfe = X.columns[rfe_features.support_].tolist()
print("Selected features using RFE method:", selected_features_rfe)
```

```python
common_features = set(selected_features_corr) & set(selected_features_kbest) & set(selected_features_rfe)
print("Common features across methods:", list(common_features))
```

```python
from sklearn.feature_selection import mutual_info_classif

# Calculate information gain (mutual information) for each feature
info_gain = mutual_info_classif(X, y)

# Create a DataFrame to store feature names and their information gain scores
info_gain_df = pd.DataFrame({'Feature': X.columns, 'InfoGain': info_gain})

# Sort features based on information gain scores
info_gain_df = info_gain_df.sort_values(by='InfoGain', ascending=False)

# Select top features based on information gain
top_features_info_gain = info_gain_df.head(5)['Feature'].tolist()  # Select top 5 features
print("Selected features using Information Gain method:", top_features_info_gain)
```
